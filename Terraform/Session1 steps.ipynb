### Installing Chocolaty and Terraform 

I chose chocolatey to install terraform you can use download installer from official site or use other package manager as per your needs.

After installing terraform using chocolatey if you run the following command to check
~~~
Terraform
~~~


### Initializing and applying Terraform 

Write a basic "tf" document to launch infrastucture via terraform and then intialize terraform using following command in same location as of your .tf file
~~~
Terraform init
~~~
